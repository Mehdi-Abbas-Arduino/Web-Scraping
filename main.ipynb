```python
from bs4 import BeautifulSoup
import requests
```

```python
r = requests.get("https://www.google.com/search?q=Breaking+Olympics&oi=ddle&ct=348046281&hl=en-GB&source=doodle-ntp&ved=0ahUKEwio56K_mueHAxXzBdsEHdX4ERkQPQgB")
with open("index.html",mode = 'w') as f :
    f.write(r.text)
```

```python
with open("index.html") as html_file:
    soup = BeautifulSoup(html_file,'lxml')
    print(soup.prettify())
```

```python
match = soup.title.text
print(match)
```
